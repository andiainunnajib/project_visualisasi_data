```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
fig = plt.figure(figsize=(4,4), dpi=300)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig = plt.figure(figsize=(1,4), dpi=300)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig, axes = plt.subplots(2,1, figsize=(4,4), dpi = 200)

axes[0].plot(x,y)
axes[0].set_title('Judul 1')

axes[1].plot(x,y)
axes[1].set_title('Judul 2')


plt.tight_layout()

plt.show()
```

```python
#untuk cetak gambar hahaha
fig.savefig('gamabarku1.png', dpi=300)
```

# TAMBAHIN LEGEND

```python
a = np.arange(1, 15, 0.1)
b = a**2

fig = plt.figure(facecolor='yellow',dpi = 200) #untuk kasih background pakai facecolor


bx = fig.add_axes([0,0,1,1])
bx.plot(a,b, label='satu')
bx.plot(a,b**3, label ='dua' )
bx.grid(True) #untuk memberikan grid
bx.legend(loc = (0.5,0.5)) #jika ingin mrngubah leta pake loc

plt.show()
```

```python
#UNTUK MEMAKAI LINEWIDHT


fig = plt.figure(facecolor='yellow',dpi = 200) #untuk kasih background pakai facecolor


bx = fig.add_axes([0,0,1,1])
bx.plot(a,b, label='satu', lw=1, ls='-')
bx.plot(a,b**3, label ='dua', lw=2, ls='--', marker='+', markersize=10, mfc='yellow', mew=3, mec='red' )
bx.plot(a,b**4, label='tiga', lw=3, ls='-.', marker='o', markersize=10, mfc='yellow', mew=3, mec='red' )
bx.grid(True) #untuk memberikan grid
bx.legend() #jika ingin mrngubah leta pake loc

#memberikan limit pada x dan y
bx.set_xlim([0,12])
bx.set_ylim([0,2000])

plt.show()
```
