```python
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y);
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.title('Contoh grafik')

plt.show()
```

```python
# buat dua grafik berdampingan

plt.subplot(121)
plt.plot(x,y, color='red')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')


plt.subplot(122)
plt.plot(y,x)
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.show()
           
           
```

```python
# buat dua grafik berdampingan

plt.subplot(121)
plt.plot(x,y, color='red')
plt.xlabel('sumbu x1')
plt.ylabel('sumbu y1')


plt.subplot(122)
plt.plot(y,x)
plt.xlabel('sumbu x2')
plt.ylabel('sumbu y2')
plt.legend(['sumubu x'])
#karen tumpah tindih tambahkan
plt.tight_layout()
plt.show()
           
           
```

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
x
```

```python
y
```

```python
# Versi matplotlib OOP

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Judul grafiknya')
plt.show()
```

```python
# Versi matplotlib OOP

fig = plt.figure()
axes = fig.add_axes([1, 2, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Judul grafiknya')
plt.show()
```

```python
# Versi matplotlib OOP

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x,y, color = 'red')
axes1.set_xlabel('Sumbu X besar')
axes1.set_ylabel('Sumbu y besar')
axes1.set_title('Judul axes 1')

axes2.plot(y,x, color = 'blue')
axes2.set_xlabel('Sumbu X kecil')
axes2.set_ylabel('Sumnbu Y kecil')
axes2.set_title('Judul axes 2')

```

```python
# versi 2
fig, axes = plt.subplots(1,2) #3 baris dan 2 kolom

# Plot untuk  axes pertama
axes[0].plot(x,y, color = 'green')
axes[0].set_xlabel('Sumbu X1')
axes[0].set_ylabel('Sumbu Y1')
axes[0].set_title('Judul 1')


#plot untuk axes kedua
axes[1].plot(x,y, color='red')
axes[1].set_xlabel('Sumbu X2')
axes[1].set_ylabel('Sumbu Y2')
axes[1].set_title('Judul 2')



#biar tidak berdempetan
fig.tight_layout()
```
